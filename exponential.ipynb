```python
 # Exponential distribution libraries
import numpy as np
import math
from scipy.stats import norm
from matplotlib import pyplot as plt
```

```python
# basic exponential function
delta_x = 0.2
x = np.arange(0, 5, delta_x)
x_exp = []
for value in x:  
    x_exp.append(math.exp(-value))
plt.plot(x, x_exp, marker='.', color='red', linewidth=3, mec='black', mfc='yellow', markersize=7)
plt.title('basic exponential function e^(-x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.ylabel('f(x)', fontsize=20)
plt.grid()
plt.show;
```

```python
# exponential distribution - three examples(can add more)
lambda_e1 = 0.5
lambda_e2 = 1.0
lambda_e3 = 1.5
lambda_list = [lambda_e1, lambda_e2, lambda_e3]

delta_x = 0.5
x_plot = np.arange(0.0, 10.0, delta_x)

for n in range(len(lambda_list)):
    p_list = []
    beta = 1 / lambda_list[n]
    for x in x_plot:
        p = (lambda_list[n] * (math.exp(-lambda_list[n] * x)))
        p_list.append(p)
    plt.plot(x_plot, p_list, label='lambda: {0:.2f}\nbeta: {1:.2f}'.format(lambda_list[n], beta)  )
plt.title('exponential distribution', fontsize=15)

#plt.plot(x_plot, p2_list, label='lambda: {0:.2f}\nbeta: {1:.2f}'.format(lambda_e2, beta_e2)  )
#plt.plot(x_plot, p3_list, label='lambda: {0:.2f}\nbeta: {1:.2f}'.format(lambda_e3, beta_e3)  )
plt.legend(title = 'parameters', loc = 'upper right')
plt.ylabel('relative probability', fontsize=12)
plt.xlabel('value', fontsize=12)
plt.grid()
plt.show();
```

```python
## exponential distribution
#normalize to 1 or sum under curve = 1 (lambda factor in equat)
lambda_e1 = 0.5
lambda_e2 = 1.0
lambda_e3 = 1.5
lambda_list = [lambda_e1, lambda_e2, lambda_e3]

delta_x = 0.5
x_range = np.arange(0.0, 15.0, delta_x)

print('approx area under curve for delta x =', delta_x, '\b:')
for n in range(len(lambda_list)):
    p_list = []
    beta = 1 / lambda_list[n]
    for x in np.arange(0.0, 15.0, delta_x):
        p = (lambda_list[n] * (math.exp(-lambda_list[n] * x)))
#        p = ((math.exp(-lambda_list[n]*x)))   #without lambda factor
        p_list.append(p)
    plt.plot(x_range, p_list, marker='.', label='lambda: {0:.2f}\nbeta: {1:.2f}'.format(lambda_list[n], beta))
# question - how can probability be > 1.0 ?
# calculate area under curves
    area_list = []
    for i in range(1, len(p_list)):
        triangle = ((p_list[i-1] - p_list[i])*delta_x)/2
        rectangle = p_list[i] * delta_x
        area = triangle + rectangle
        area_list.append(area)
    print('  area under lambda={:.2f} {:.3f} '.format(lambda_list[n], sum(area_list)))
# answer - As long as total cumulative probability equals 1.0, it's all relative to actual probability
# if the curve is normalized to start at 1.0, the area under curve becomes beta

plt.title('exponential distribution', fontsize=15);
plt.legend(title = 'parameters', loc = 'upper right')
plt.ylabel('relative probability', fontsize=12)
plt.xlabel('value', fontsize=12)
plt.grid()
plt.show();
```

```python
# CDF for exponential distribution
delta_x = 0.5
x_range = np.arange(0.0, 15.0, delta_x)

for n in range(len(lambda_list)):
    cdf_list = []
    beta = 1 / lambda_list[n]
    for x in x_range:
        cdf = 1 - math.exp(-lambda_list[n]*x)
        cdf_list.append(cdf)
    plt.plot(x_range, cdf_list, marker='.', label='lambda: {0:.2f}\nbeta: {1:.2f}'.format(lambda_list[n], beta))

plt.title('exponential cumulative probability', fontsize=15)
plt.legend(title = 'parameters', loc = 'lower right')
plt.ylabel('probability', fontsize=12)
plt.xlabel('value', fontsize=12)
plt.grid()
plt.show();

# PPF functionfor n in range(len(lambda_list)):
for n in range(len(lambda_list)):
    cdf_list = []
    beta = 1 / lambda_list[n]
    for x in x_range:
        cdf = 1 - math.exp(-lambda_list[n]*x)
        cdf_list.append(cdf)
    plt.plot(cdf_list, x_range, marker='.', label='lambda: {0:.2f}\nbeta: {1:.2f}'.format(lambda_list[n], beta))
    
plt.title('Probability Percent Function', fontsize=15)
plt.legend(title = 'parameters', loc = 'upper left')
plt.xlabel('cumulative probability', fontsize=12)
plt.ylabel('value', fontsize=12)
plt.grid()
plt.show();
```

```python
# try a horizontal shift with basic math
lambda_e = 1.5
beta_e = 1/lambda_e1
delta_x = 0.1
ld_list1 = []
ld_list2 = []
x_list = []
right_shift = 2

x_range = np.arange(0.0, 10.0, delta_x)
for x in np.arange(0.0, 10.0, delta_x):
    e_exp = (x - right_shift)
    lambda_dist1 = lambda_e*(math.exp(-x*lambda_e))
    lambda_dist2 = lambda_e*(math.exp(-(e_exp)*(lambda_e)))
    if e_exp  < 0:
        lambda_dist2 = 0
    ld_list1.append(lambda_dist1)
    ld_list2.append(lambda_dist2)
    x_list.append(x)
    
plt.title('exponential distribution - math', fontsize=15)
plt.plot(x_list, ld_list1, marker='.', label='lambda: {0:.2f}\nbeta: {1:.2f}'.format(lambda_e, beta_e))
plt.plot(x_list, ld_list2, marker='.', label='lambda: {0:.2f}\nbeta: {1:.2f}, shift: {2:.2f}'\
         .format(lambda_e, beta_e, right_shift))
plt.legend(title = 'parameters', loc = 'upper right')
plt.ylabel('relative probability', fontsize=12)
plt.xlabel('value', fontsize=12)
plt.grid()
plt.show();

# using scipy stats
from scipy.stats import expon
delta_x = 0.1
x_values = [value for value in np.arange(0, 8, delta_x)]

# Varying arguments
scale1 = 0.8
scale2 = 1.5
loc1 = 0
loc2 = 3
dist1 = expon(loc=loc1, scale=scale1)
dist2 = expon(loc=loc2, scale=scale2)
y1 = dist1.pdf(x_values)
y2 = dist2.pdf(x_values)

plt.plot(x_values, y1, '.', label='y1 beta: {0:.2f}, loc:{1:.2f}, mean: {2:.2f}, st dev:{3:.2f}'\
         .format(scale1, loc1, dist1.mean(), dist1.std()))
plt.plot(x_values, y2, '.', label='y2 beta: {0:.2f}, loc:{1:.2f}, mean: {2:.2f}, std dev:{3:.2f}'\
         .format(scale2, loc2, dist2.mean(), dist2.std()))
plt.legend(loc='upper right')
plt.title('exponential distribution - scipy', fontsize=15)
plt.ylabel('relative probability', fontsize=12)
plt.xlabel('value', fontsize=12)
plt.grid()
plt.show();
```

```python
# sample an exponential distribution
from numpy.random import exponential
# define the distribution
beta1 = 1
beta2 = 3
n = 1000
nn = np.arange(0, n, 1)

# generate the sample
sample1A = exponential(beta1, n)
sample1B = exponential(beta1, n)
sample2A = exponential(beta2, n)
sample2B = exponential(beta2, n)

# statistics function
def sample_stat(sample):
    mean = sum(sample)/len(sample)
    dif_square_list = []
    for value in sample:
        dif_square = (mean - value)**2
        dif_square_list.append(dif_square)
    sum_squares = sum(dif_square_list)
    variance = sum_squares/(len(sample))
    return mean, variance

# statistics
mean1A, variance1A = sample_stat(sample1A)
mean1B, variance1B = sample_stat(sample1B)
mean2A, variance2A = sample_stat(sample2A)
mean2B, variance2B = sample_stat(sample2B)
print('means:{0:.4f}, {1:.4f}, {2:.4f}, {3:.4f}'.format(mean1A, mean1B, mean2A, mean2B))
print('variances:{0:.4f}, {1:.4f}, {2:.4f}, {3:.4f}'.format(variance1A, variance1B, variance2A, variance2B))

# using numpy statistics
print('sample1A: maximum= {0:.3f}  mean= {1:.3f}  variance= {2:.3f}  std= {3:.3f}'\
      .format(max(sample1A), np.mean(sample1A), np.var(sample1A), math.sqrt(np.var(sample1A))))
print('sample1B: maximum= {0:.3f}  mean= {1:.3f}  variance= {2:.3f}  std= {3:.3f}'\
      .format(max(sample1B), np.mean(sample1B), np.var(sample1B), math.sqrt(np.var(sample1B))))
print('sample2A: maximum= {0:.3f}  mean= {1:.3f}  variance= {2:.3f}  std= {3:.3f}'\
      .format(max(sample2A), np.mean(sample2A), np.var(sample2A), math.sqrt(np.var(sample2A))))
print('sample2B: maximum= {0:.3f}  mean= {1:.3f}  variance= {2:.3f}  std= {3:.3f}'\
      .format(max(sample2B), np.mean(sample2B), np.var(sample2B), math.sqrt(np.var(sample2B))))


# plot sample 1A
plt.scatter(nn, sample1A, marker='.', label='sample 1A as selected')
plt.scatter(nn, sorted(sample1A), marker='.', label='sample 1A sorted from {0:.2f} to {1:.2f}' \
            .format(min(sample1A), max(sample1A)))
plt.title('Exponential Distribution {} samples  beta: {:.2f}'.format(n, beta1))
plt.xlabel('sample number', fontsize=12)
plt.ylabel('value', fontsize=12)
plt.legend()
plt.show();

# plot to compare samples
plt.plot(sorted(sample1A), label='sample 1A beta: {:.2f}'.format(beta1))
plt.plot(sorted(sample1B), label='sample 1B beta: {:.2f}'.format(beta1))
plt.plot(sorted(sample2A), label='sample 2A beta: {:.2f}'.format(beta2))
plt.plot(sorted(sample2B), label='sample 2B beta: {:.2f}'.format(beta2))
plt.scatter(n,max(sample1A), label='sample 1A maximum {:.2f}'.format(max(sample1A)))
plt.scatter(n,max(sample1B), label='sample 1B maximum {:.2f}'.format(max(sample1B)))
plt.scatter(n,max(sample2A), label='sample 2A maximum {:.2f}'.format(max(sample2A)))
plt.scatter(n,max(sample2B), label='sample 2B maximum {:.2f}'.format(max(sample2B)))
plt.legend(loc='upper left')
plt.title('Exponential Sampling {} samples - sorted'.format(n))
plt.xlabel('sample number', fontsize=12)
plt.ylabel('value', fontsize=12)
plt.show();

#draw histogram
binwidth = 1
# counts[0]/35 is used to normalize the min/max markers on the x axis
counts, bins, bars = plt.hist(sample1A, bins=range(math.floor(min(sample1A)), math.ceil(max(sample1A)) + binwidth, binwidth),\
            label='mean= {0:.3f}  std= {1:.3f}'.format(mean1A, math.sqrt(variance1A)))
plt.title('Exponential Distribution Sampling \nsample 1A  beta={0:.2f}  {1:}  samples'.format(beta1, n), fontsize=15)
plt.scatter(min(sample1A), counts[0]/35, color='red', marker='|', s=200, zorder=2,\
            label='minimum={:.2f}'.format(min(sample1A)))
plt.scatter(max(sample1A), counts[0]/35, color='red', marker='|', s=200, zorder=2,\
            label='maximum={:.2f}'.format(max(sample1A)))
plt.xlabel('value', fontsize=12)
plt.ylabel('occurences', fontsize=12)
plt.legend()
plt.show();

counts, bins, bars = plt.hist(sample1B, bins=range(math.floor(min(sample1B)), math.ceil(max(sample1B)) + binwidth, binwidth),\
            label='mean= {0:.3f}  std= {1:.3f}'.format(mean1B, math.sqrt(variance1B)))
plt.title('Exponential Distribution Sampling \nsample 1B  beta={0:.2f}  {1:}  samples'.format(beta1, n), fontsize=15)
plt.scatter(min(sample1B), counts[0]/35, color='red', marker='|', s=200, zorder=2,\
            label='minimum={:.2f}'.format(min(sample1B)))
plt.scatter(max(sample1B), counts[0]/35, color='red', marker='|', s=200, zorder=2,\
            label='maximum={:.2f}'.format(max(sample1B)))
plt.xlabel('value', fontsize=12)
plt.ylabel('occurences', fontsize=12)
plt.legend()
plt.show();

counts, bins, bars = plt.hist(sample2A, bins=range(math.floor(min(sample2A)), math.ceil(max(sample2A)) + binwidth, binwidth),\
            label='mean= {0:.3f}  std= {1:.3f}'.format(mean2A, math.sqrt(variance2A)))
plt.title('Exponential Distribution Sampling \nsample 2A  beta={0:.2f}  {1:}  samples'.format(beta2, n), fontsize=15)
plt.scatter(min(sample2A), counts[0]/35, color='red', marker='|', s=200, zorder=2,\
            label='minimum={:.2f}'.format(min(sample2A)))
plt.scatter(max(sample2A), counts[0]/35, color='red', marker='|', s=200, zorder=2,\
            label='maximum={:.2f}'.format(max(sample2A)))
plt.xlabel('value', fontsize=12)
plt.ylabel('occurences', fontsize=12)
plt.legend()
plt.show();

counts, bins, bars = plt.hist(sample2B, bins=range(math.floor(min(sample2B)), math.ceil(max(sample2B)) + binwidth, binwidth),\
            label='mean= {0:.3f}  std= {1:.3f}'.format(mean2B, math.sqrt(variance2B)))
plt.title('Exponential Distribution Sampling \nsample 2B  beta={0:.2f}  {1:}  samples'.format(beta2, n), fontsize=15)
plt.scatter(min(sample2B), counts[0]/35, color='red', marker='|', s=200, zorder=2,\
            label='minimum={:.2f}'.format(min(sample2B)))
plt.scatter(max(sample2B), counts[0]/35, color='red', marker='|', s=200, zorder=2,\
            label='maximum={:.2f}'.format(max(sample2B)))
plt.xlabel('value', fontsize=12)
plt.ylabel('occurences', fontsize=12)
plt.legend()
plt.show();
```
