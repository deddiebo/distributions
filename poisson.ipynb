```python
 # poisson distribution
import numpy as np
import math
from scipy.stats import poisson
from matplotlib import pyplot as plt
```

```python
# using math function - single plot - PMF, CDF, PPF
# define distribution parameters
lambda_val = 6
 
# create distribution
k_values = [value for value in range(0, 21)]
p_value = [((lambda_val**k)/math.factorial(k))*math.exp(-lambda_val) for k in k_values]

# plot values
plt.plot(k_values, p_value, marker='o', label='lambda: {:}'.format(lambda_val))
plt.xlabel('number of occurences', fontsize=12)
plt.ylabel('probability density', fontsize=12)
plt.title('poisson distribution', fontsize=15)
plt.legend(title = 'parameter', loc = 'upper right')
plt.grid()
plt.show();

# plot CDF
cum_prob = [poisson.cdf(k, lambda_val) for k in k_values]
plt.plot(k_values, cum_prob, marker='.', label=chr(955)+' : '+str(lambda_val))
plt.xlabel('number of occurences', fontsize=12)
plt.ylabel('probability', fontsize=12)
plt.title("CDF of poisson distribution", fontsize=15)
plt.legend(title = 'parameters', loc = 'center right')
plt.grid()
plt.show();

# plot PPF
plt.plot(cum_prob, k_values, marker='.', label=chr(955)+' : '+str(lambda_val))
plt.ylabel('number of occurences', fontsize=12)
plt.xlabel('probability', fontsize=12)
plt.title("PPF of poisson distribution", fontsize=15)
plt.legend(title = 'parameters', loc = 'upper left')
plt.grid()
plt.show();
```

```python
# look up value
print('poisson distribution - lambda value:', lambda_val )
value = 7
print('   probability of {0:} occurences = {1:.3f}'.format(value, p_value[value]))
print('   cumulative probability of {0:} occurences = {1:.3f}'.format(value, cum_prob[value]))
```

```python
# using scipy CMF, CDF, PPF
# define distribution parameters - plot 3 or more values
lambda_val1 = 0.7
lambda_val2 = 3
lambda_val3 = 10
lambda_val = [lambda_val1, lambda_val2, lambda_val3]

for n in range(len(lambda_val)):
    # create distribution
    dist = poisson(lambda_val[n])
    # statistics
    mean, var = poisson.stats(lambda_val[n], moments='mv')

# plot PMF
    values = [value for value in range(0, 21)]
    probabilities = [dist.pmf(value) for value in values]
    plt.plot(values, probabilities, marker='o', label='lambda: {0:}  mean: {1:}  var: {1:}'\
         .format(lambda_val[n], mean, var))
plt.xlabel('number of occurences', fontsize=12)
plt.ylabel('probability density', fontsize=12)
plt.title('poisson distribution', fontsize=15)
plt.legend(title ='parameters', loc ='upper right')
plt.grid()
plt.show();

# plot CDF
for n in range(len(lambda_val)):
    cum_prob = [poisson.cdf(k, lambda_val[n]) for k in k_values]
    plt.plot(values, cum_prob, marker='.', label=chr(955)+' : '+str(lambda_val[n]))
plt.xlabel('number of occurences', fontsize=12)
plt.ylabel('probability', fontsize=12)
plt.title("CDF of poisson distribution", fontsize=15)
plt.legend(title = 'parameters', loc = 'center right')
plt.grid()
plt.show();

# plot PPF
for n in range(len(lambda_val)):
    cum_prob = [poisson.cdf(k, lambda_val[n]) for k in k_values]
    plt.plot(cum_prob, values, marker='.', label=chr(955)+' : '+str(lambda_val[n]))
#plt.plot(cum_prob2, values, marker='.', label=chr(955)+' : '+str(lambda_val2))
#plt.plot(cum_prob3, values, marker='.', label=chr(955)+' : '+str(lambda_val3))
plt.ylabel('number of occurences', fontsize=12)
plt.xlabel('probability', fontsize=12)
plt.title("PPF of poisson distribution", fontsize=15)
plt.legend(title = 'parameters', loc = 'upper left')
plt.grid()
plt.show();
```

```python
# generate poisson distribution from scipy
lambda_val = 10
samples = 5000
data = poisson.rvs(lambda_val, size=samples)

# plot data
plt.scatter(range(samples), data, marker='.')
plt.ylabel('value', fontsize=12)
plt.xlabel('sample number', fontsize=12)
plt.title('Poisson Distribution {0:} samples \n  lambda: {1:}'.format(samples, lambda_val), fontsize=15)
plt.show();

# statistics
len_data = len(data)
mean = sum(data)/len(data)
max_data = max(data)
var_data = [((mean - data[n])**2) for n in range(len_data)]
variance = sum(var_data) / len_data

# histogram
plt.hist(data, bins= range(0, max_data+1, 1), label='maximum: {0:}\nmean: {1:.2f}\nvariance: {2:.2f}'\
         .format(max_data, mean, variance))
plt.ylabel('number of occurences', fontsize=12)
plt.xlabel('value\nexpected: mean= {0:.2f}  variance= {1:.2f}'\
           .format(lambda_val, lambda_val), fontsize=12)
plt.title('histogram of distribution\n{0:} samples   lambda: {1:}'.format(samples, lambda_val) , fontsize=15)
plt.legend()
plt.show();
```
